```python
x = iris.drop('species', axis=1)
y = iris['species']
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
gb = GaussianNB()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
gb.fit(x_train,y_train)
```

```python
y_predict = gb.predict(x_test)
y_predict
```

```python
print("Confusion Matrix :")
sns.heatmap(confusion_matrix(y_test,y_predict),annot=True,cmap="Blues",fmt="d")
```

```python
print("Classification Report :")
print(classification_report(y_test,y_predict))
print("\n accuracy score =",np.round(accuracy_score(y_test,y_predict),3)*100,"%")
```

```python
from sklearn.svm import SVC
svm = SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
y_predict_svm = svm.predict(x_test)
y_predict_svm
```

```python
print("Confusion Matrix :")
sns.heatmap(confusion_matrix(y_test,y_predict_svm),annot=True,cmap="Blues",fmt="d")
```

```python
print("Classification Report :")
print(classification_report(y_test,y_predict_svm))
print("\n accuracy score =",np.round(accuracy_score(y_test,y_predict_svm),3)*100,"%")
```
